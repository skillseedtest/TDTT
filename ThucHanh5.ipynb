```python
#W5A1
def Max_Of_Two(a: int, b: int) -> int:
    return b if a < b else a
a, b = map(int, input().split())
print(Max_Of_Two(a, b))
```

```python
#W5A2
def swap(a, b):
    return (b, a)
a, b = map(int, input().split())
print(swap(a, b))
```

```python
#W5A3
def isPrime(n):
    i = 2
    while i * i <= n:
        if n % i == 0:
            return False
        i += 1
    return n >= 2
print(isPrime(int(input())))
```

```python
#W5A4
def isPerfect(n):
    return sum(i * (n % i == 0) + n // i * (n % i == 0 and i != n // i) for i in range(1, int(n ** .5) + 1)) == (n << 1)
print(isPerfect(int(input())))
```

```python
#W5A5
def firstFind(a, n):
    for i in range(len(a)):
        if a[i] == n:
            return i + 1
    return -1
print(firstFind(list(map(int, input().split())), int(input())))
```

```python
#W5A6
def fact(n):
    return fact(n - 1) * n if n else 1
print(fact(int(input())))
```

```python
#W5A7
def conv(s):
    return float(s) if s != '.' else 0
def calc(s):
    sgn_a, sgn_b, sa, sb, ls_op = 1, 1, '', '', ['+', '-', '*', '/']
    for i in range(len(s)):
        if i and s[i] in ls_op and s[i - 1] not in ls_op:
            op, sa, sb, sgn_a, sgn_b = s[i], sb, sa, sgn_b, sgn_a
        elif s[i] not in ls_op:
            sa += s[i]
        elif s[i] == '-':
            sgn_a *= -1
    a, b = conv(sb) * sgn_b, conv(sa) * sgn_a
    return a + b if op == '+' else a - b if op == '-' else a * b if op == '*' else a / b
print(f"{calc(''.join(input().split())):.2f}")
```

```python
#W5A8
def distHamming(a, b):
    ans = 0
    while max(a, b) > 0:
        ans += (a ^ b) & 1
        a >>= 1
        b >>= 1
    return ans
a, b = map(int, input().split())
print(distHamming(a, b))
```

```python
#W5A9
def digSum(n):
    return sum(int(i) for i in n)
print(digSum(input()))
```

```python
#W5A10
def don_anh(a, b):
    ls = ['?'] * 26
    for i in range(len(a)):
        j = ord(a[i]) - ord('a')
        if ls[j] == '?':
            ls[j] = b[i]
        elif ls[j] != b[i]:
            return False
    return True
def dang_cau(a, b):
    return don_anh(a, b) and don_anh(b, a)
a, b = input().split()
print(dang_cau(a, b))
```
